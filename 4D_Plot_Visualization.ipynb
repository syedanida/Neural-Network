```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data with 3 variables
def generate_data(n_samples=500):
    # Generate x, y, z as random values between -5 and 5
    x = np.random.uniform(-5, 5, n_samples)
    y = np.random.uniform(-5, 5, n_samples)
    z = np.random.uniform(-5, 5, n_samples)

    # Create a non-linear equation for the fourth dimension (w)
    # w = sin(x) + cos(y) * exp(z/10)
    w = np.sin(x) + np.cos(y) * np.exp(z/10)

    return x, y, z, w

# Generate the data
x, y, z, w = generate_data(1000)

# Create a 4D scatter plot (using color as the 4th dimension)
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Create scatter plot with color representing the 4th dimension (w)
scatter = ax.scatter(x, y, z, c=w, cmap='viridis', s=30, alpha=0.8)

# Add a color bar to show the scale of the 4th dimension
cbar = plt.colorbar(scatter)
cbar.set_label('w = sin(x) + cos(y) * exp(z/10)', rotation=270, labelpad=20)

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('4D Visualization: 3D Scatter Plot with Color as 4th Dimension')

# Add grid lines for better visibility
ax.grid(True)

# Adjust the viewing angle
ax.view_init(elev=30, azim=45)

# Show the plot
plt.tight_layout()
plt.show()

# Let's also create an interactive rotation function (uncomment to use in interactive environments)
"""
for angle in range(0, 360, 5):
    ax.view_init(30, angle)
    plt.draw()
    plt.pause(0.1)
"""
```
